# Introductory Examples

#### Importing Main Libraries

```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

## Obtain Data

```python
# import usa.gov bitly dataset
# NB: use appropriate pathname on your computer
path = '../../Datasets/usagov_bitly_data2012-03-16-1331923249.txt'

# read the first line to get a sense of data
open(path).readline()

# convert json object into Python dictionary
# first import json library
import json

# test with just the first line in file
json.loads(open(path).readline())

# convert all json objects to python dict
# use list comprehension
records = [ json.loads(line) for line in open(path)]

# check first item in records list
records[0]

# access time zone in first item 
records[0]['tz']
```

## Data Scrubing with Pandas

```python
# store list of dicts object as pandas dataframe
frame = DataFrame(records)

# check out dataframe
frame

# first 10 time zones in data frame
frame['tz'][:10]

# total counts of time zones in data frame
tz_counts = frame['tz'].value_counts()

# first 10 rows of tz_counts Series
tz_counts[:10]

# replace NA values with Missing using fillna method
clean_tz = frame['tz'].fillna('Missing')

# replace empty data with 'Unknown' value
clean_tz[clean_tz == ''] = 'Unknown'

tz_counts = clean_tz.value_counts()

tz_counts[:10]
```

## Explore Data

#### Browser Capability

```python
# plot (at bottom) first 10 time zones on a horizontal barplot
tz_counts[:10].plot(kind='barh', rot=0)

# obtain the browser capability from the agent string
results = Series([x.split()[0] for x in frame.a.dropna()])
results[:5]

# get counts for first 8 different types of browser capability in results series
results.value_counts()[:10]
```

#### Windows and Non-Windows Users

```python
# find number of Windows and non-Windows users in top time zones
cframe = frame[frame.a.notnull()]

operating_system = np.where(cframe['a'].str.contains('Windows'),
                            'Windows', 'Not Windows')
operating_system[:5]

# Group data by its time zone column and operating system
by_tz_os = cframe.groupby(['tz', operating_system])

agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]

# select top overall time zones

indexer = agg_counts.sum(1).argsort()
indexer[:10]

count_subset = agg_counts.take(indexer)[-10:]

count_subset

count_subset.plot(kind='barh', stacked=True)

normed_subset = count_subset.div(count_subset.sum(1), axis=0)

normed_subset.plot(kind='barh', stacked=True)
```

## MovieLens 1M Data Set

### Importing Data

```python
# save movielen folder name
movielen_folder = '../../Datasets/ml-1m/'

# import user information data as data frame
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
user_data = movielen_folder+'users.dat'
users = pd.read_table(user_data, sep='::', 
                      header=None, names=unames)

# import ratings data as data frame
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
rating_data = movielen_folder+'ratings.dat'
ratings = pd.read_table(rating_data, sep='::', header=None,
                        names=rnames)

# import movies data as data frame
mnames = ['movie_id', 'title', 'genres']
movie_data = movielen_folder+'movies.dat'
movies = pd.read_table(movie_data, sep='::', header=None,
                       names=mnames)
```

```python
users[:5]

ratings[:5]

movies[:5]

# merge data
data = pd.merge(pd.merge(ratings, users), movies)

data.ix[0]

mean_ratings = data.pivot_table('rating', rows='title', cols='gender',
                                aggfunc='mean')

mean_ratings[:5]

# get the number of movies for each title
ratings_by_title = data.groupby('title').size()

# show first 10 movies
ratings_by_title[:10]

active_titles = ratings_by_title.index[ratings_by_title >= 250] 

active_titles

mean_ratings = mean_ratings.ix[active_titles]

mean_ratings
```

```python
# top films among female viewers
top_female_ratings = mean_ratings.sort_index(by='F', ascending=False)

top_female_ratings[:10]

# add column with difference in movie ratings for male and female
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']

sorted_by_diff = mean_ratings.sort_index(by='diff')

sorted_by_diff[:15]

# reverse order of rows
sorted_by_diff[::-1][:15]
```

### US Baby Names 1880-2010

```python
import pandas as pd

# read in 1880 names
names1880 = pd.read_csv('../../Datasets/names/yob1880.txt', 
                        names=['name', 'sex', 'births'])

names1880

names1880.groupby('sex').births.sum()

# 2013 is the last available year right now
years = range(1880, 2014)

pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = '../../Datasets/names/yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    
    frame['year'] = year
    pieces.append(frame)
    
    # Concatenate everything into a single DataFrame
    names = pd.concat(pieces, ignore_index=True)

    

names
```

```python
# aggregate data at the year and sex level
total_births = names.pivot_table('births', rows='year', cols='sex',
                                 aggfunc=sum)

total_births.tail()

total_births.plot(title='Total births by sex and year')
```

```python
# create function to add 'prop' column whihc shows fraction of babies
# given each name relative to the total number of births

def add_prop(group):
    #  Integer divisions floors
    births = group.births.astype(float)
    
    group['prop'] = births/ births.sum()
    return group
names = names.groupby(['year', 'sex']).apply(add_prop)

names
```

```python
# sanity check
np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)

# function to extract top 1000 names for each sex/year combinatin

def get_top1000(group):
    return group.sort_index(by='births', ascending=False)[:1000]

grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)

top1000
```

### Analyzing Naming Trends

```python
boys = top1000[top1000.sex == 'M']

girls = top1000[top1000.sex == 'F']

# pivot table of total number of births by year and name
total_births = top1000.pivot_table('births', rows='year', cols='name',
                                   aggfunc=sum)

total_births

subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]

subset.plot(subplots=True, figsize=(12, 10), grid=False,
            title = "Number of birthds per year")
```
